```python
method_dict = dict()
temp = data[data['Have you studied data visualization in school (or other formal environment) or did you learn how to do it on your own?'] == 'Mostly From School (or other formal courses)']
temp.head()
```

```python
for index, row in temp[['What technologies do you use to visualize data? Select all that apply.']].iterrows():
    arr = row.values[0]
    arr = arr.split(', ')
#     print(arr)
    for method in arr:
        if method not in method_dict:
            method_dict[method] = 1
        else:
            method_dict[method] += 1
```
